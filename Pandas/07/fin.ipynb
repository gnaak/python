```python
import pandas as pd
# 데이터 시각화 
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 글씨체
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False # 음수 사용
# 연령별 투자 성향 분석:
# 각 연령 그룹이 어떤 종류의 투자를 선호하며, 그 선호도에 어떤 차이가 있는지 확인합니다. 예를 들어, 어린 세대는 주식 시장에 더 많이 투자하는 경향이 있는지, 중장년층은 안전한 투자 수단을 선호하는지 등을 분석합니다.

# 성별에 따른 투자 성향 비교:
# 남성과 여성 간에는 어떤 투자 성향의 차이가 있는지 확인합니다. 특정 투자 옵션에 대한 관심이나 선호도에서 성별에 따른 특징을 분석할 수 있습니다.

# 연령/성별에 따른 이유 분석:
# 각 투자 옵션을 선택하는 이유에 대한 설문조사 결과를 바탕으로, 연령별 또는 성별별로 어떤 이유가 주요하게 나타나는지 확인합니다. 예를 들어, 젊은 세대는 수익을 높이기 위해 투자하는 경우가 많은지 등을 파악합니다.

# 다양한 시각화 기법 활용:
# 성별과 연령에 따른 투자 성향을 시각적으로 보여주는 다양한 차트 및 그래픽을 사용합니다. 막대 그래프, 원 그래프, 히트맵 등을 활용하여 효과적인 시각화를 구현합니다.

# 통계적 검증:
# 통계 분석을 통해 연령과 성별이 투자 성향에 미치는 영향의 유의성을 확인합니다. 투자 성향에 영향을 미치는 요인을 식별하고, 각 요인 간의 관계를 분석합니다.
```

```python
df = pd.read_csv('Finance_data.csv')
df.drop(columns=['Mutual_Funds', 'Equity_Market' , 'Debentures', 'Government_Bonds', 'Fixed_Deposits' ,'PPF', 'Gold'])
```

```python
# 1. 연령별 투자 성향 분석 

bins = [20, 30, 40]  # 나이대 구간 설정
labels = ['20대', '30대']  # 구간별 라벨 설정

df['Age_Group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)  # cut 함수를 사용하여 나이대 구간 (right=False면 10이상 20미만 )
df.drop(columns='age', inplace=True)

age_grouped = df.groupby('Age_Group')
```

```python
for age_group, group_data in age_grouped:
    print(f"\nAge Group: {age_group}")
    print(group_data['Investment_Avenues'].value_counts())
```
