```python
# 축
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 글씨체
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False # 음수 사용

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family' : 'HYGungSo-Bold', 'size' : 20})
plt.xlabel('x축', color = 'red')
plt.ylabel('y축', color = 'green')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family' : 'HYGungSo-Bold', 'size' : 20})
plt.xlabel('x축', color = 'red', loc = 'right') # left, center, right
plt.ylabel('y축', color = 'green', loc = 'top') # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
