```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 글씨체
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False # 음수 사용
# matplotlib marker style
# matplotlib line style
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

```python
# 마커 스타일
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v')
```

```python
plt.plot(x,y, marker='v', markersize=10)
```

```python
plt.plot(x,y, marker='x', markersize=10)
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor = 'red')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor = 'red', markerfacecolor='yellow')
```

```python
# 선 스타일
plt.plot(x,y, linestyle = ':')
```

```python
plt.plot(x,y, linestyle = '--')
```

```python
plt.plot(x,y, linestyle = '-.')
```

```python
# 라인 색
plt.plot(x,y , color='pink')
```

```python
# 포맷
plt.plot(x, y,'ro--') # color, marker, line
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x,y,'go')
```

```python
## 축약어
plt.plot(x,y,marker='o', mfc ='red', ms=10, mec='yellow', ls=':')
```
